```python
import os
import glob
import sys



DIR = os.path.abspath('')
DIR_ARCHIVES = os.path.join(DIR, '..', 'sources', 'archives')
DIR_MODULE = os.path.join(DIR, '..')

sys.path.append(DIR_MODULE)
```

```python
import utils_package_json

all = []
# loop through all directory under archives directory
for d in os.listdir(DIR_ARCHIVES):
    path = os.path.join(DIR_ARCHIVES, d)
    if os.path.isdir(path):
        # get all projects in the directory
        projects = utils_package_json.locate_projects(path)
        all.extend(projects)

# list all projects
all
```

```python
# list styled-components files
import mod_styled
all_files = []

for p in all:
  path = p['path']
  print(path)
  files = mod_styled.list_styled_components_files(path)
  all_files.extend(files)

len(all_files)

```
